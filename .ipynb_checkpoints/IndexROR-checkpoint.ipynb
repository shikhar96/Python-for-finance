```python
#Importing packages
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas_datareader import data as wb
```

```python
tickers = ['^GSPC','^FTSE','^NSEI','^GDAXI']
indexdata = pd.DataFrame()
for t in tickers:
    indexdata[t] = wb.DataReader(t, data_source='yahoo', start='2000-1-1')['Adj Close']
```

```python
indexdata.head()
```

```python
indexdata.tail()
```

```python
#Calculating simple returns of the indices
returns = (indexdata/indexdata.shift(-1)) - 1
print(returns)
```

```python
annual_returns = returns.mean() * 250
print(annual_returns)
```
